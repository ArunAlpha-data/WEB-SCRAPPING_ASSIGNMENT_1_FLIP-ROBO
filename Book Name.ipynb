# OBJECTIVE:

To write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’


# IMPORTING LIBRARIES

```python
from bs4 import BeautifulSoup
import requests
```

# REQUESTING PAGE FOR WEB SCRAPPING

```python
page=requests.get("https://bookpage.com/reviews")
page
```

# FETCHING THE PAGE CONTENTS

```python
page.content
```

# STREAMLINING THE WEBPAGE CONTENTS

```python
soup=BeautifulSoup(page.content,"html.parser")
soup
```

# FETCHING THE BOOK TITLE

```python
title=soup.find_all('h4',class_='italic') 
title
```

```python
book=[]
for i in title:
    book.append(i.a.text)
book
```

# FETCHING THE BOOK AUTHOR

```python
author=soup.find_all('p',class_='sans bold') 
author
```

```python
Author=[]
for i in author:
    Author.append(i.text.replace("\n",""))
Author
```

# FETCHING THE BOOK REVIEW

```python
rev=soup.find_all('p',class_='excerpt') 
rev
```

```python
review=[]
for i in rev:
    review.append(i.text.replace("\n",""))
review
```

# FETCHING THE BOOK GENRE

```python
genre=soup.find_all('p',class_='genre-links hidden-phone') 
genre
```

```python
Genre=[]
for i in genre:
    Genre.append(i.text.replace("\n",""))
Genre
```

# COMMITING ALL THE EXTRACTED DATA'S INTO DATA-SET

```python
import pandas as pd
BOOK_PAGE=pd.DataFrame({})
BOOK_PAGE["Book_Title"]=book[0:5]
BOOK_PAGE["Author"]=Author[0:5]
BOOK_PAGE["Genre"]=Genre[0:5]
BOOK_PAGE["Review"]=review[0:5]
BOOK_PAGE
```
